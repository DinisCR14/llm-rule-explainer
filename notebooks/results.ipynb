# One-Shot Prompting

```python
def plot_os_sim(os_files, labels, title='GPT-4o'):
    """Plot mean ROUGE-1 F1 scores for one-shot similarity experiments.

    The first label is rendered with lighter colours to visually separate it
    from the comparison methods.

    Parameters
    ----------
    os_files : list of str
        Paths to JSON result files, one per label.
    labels : list of str
        Display labels corresponding to each file.
    title : str
        Plot title string.
    """
    plot_data = []

    for label, file in zip(labels, os_files):
        with open(file, 'r') as f:
            data = json.load(f)

        rouge1_name_scores = []
        rouge1_description_scores = []

        for rule, details in data.items():
            name_scores = details['Name ROUGE Scores']['rouge1']
            description_scores = details['Description ROUGE Scores']['rouge1']
            if name_scores:
                rouge1_name_scores.append(name_scores[2])       # F1 at index 2
            if description_scores:
                rouge1_description_scores.append(description_scores[2])

        if rouge1_name_scores:
            plot_data.append((label, "Names", rouge1_name_scores))
        if rouge1_description_scores:
            plot_data.append((label, "Descriptions", rouge1_description_scores))

    data_long = {"Label": [], "Type": [], "Score": []}
    for label, score_type, scores in plot_data:
        data_long["Label"].extend([label] * len(scores))
        data_long["Type"].extend([score_type] * len(scores))
        data_long["Score"].extend(scores)

    plot_df = pd.DataFrame(data_long)
    first_label_df = plot_df[plot_df["Label"] == labels[0]]
    other_labels_df = plot_df[plot_df["Label"] != labels[0]]

    plt.figure(figsize=(10, 7))
    sns.barplot(
        data=first_label_df, x="Label", y="Score", hue="Type",
        errorbar=("ci", 95), palette={"Names": "#66CDAA", "Descriptions": "#FFA07A"},
        dodge=True, legend=False,
    )
    sns.barplot(
        data=other_labels_df, x="Label", y="Score", hue="Type",
        errorbar=("ci", 95), palette={"Names": "mediumseagreen", "Descriptions": "darkorange"},
        dodge=True, legend=True,
    )
    plt.ylabel('Mean ROUGE-1 F1 Scores', fontsize=20)
    plt.xlabel("")
    plt.title(f"{title}", fontsize=24)
    plt.ylim(0, 1)
    plt.grid(axis='y')
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=20)
    plt.legend(fontsize=20, title_fontsize=20)
    plt.tight_layout()
    plt.show()
```

```python
# File paths and corresponding labels
os_files = [
    "Results/gpt-4-turbo/os_sim_gpt-4-turbo_temp0.25_usebertFalse.json",
    "Results/gpt-4-turbo/os_sim_gpt-4-turbo_temp0.25.json",
    "Results/gpt-4-turbo/os_sim_gpt-4-turbo_temp0.25_usebertTrue.json"
]
labels = ['OS Sim', 'OS Sim Order', 'OS BERT']

plot_os_sim(os_files, labels, title='GPT-4 Turbo')
```

```python
def process_json_manual_metric(file_path, variable_keywords):
    """
    Process a JSON file to extract rules and generated descriptions, then
    applies a comparison function to each rule-description pair.

    Args:
        file_path (str): Path to the JSON file.
        variable_keywords (list): Keywords for identifying variables in the description.

    Returns:
        tuple: A list of dictionaries containing comparison results for each rule,
               and two lists containing the `recall` and `f1_score` values.
    """
    with open(file_path, 'r') as file:
        data = json.load(file)

    results = []
    recalls = []
    f1_scores = []
    
    for rule, content in data.items():
        description = content.get("Generated Description")  # Use .get() to avoid KeyError
        if description:  # Skip if description is None or empty
            result = compare_rule_and_description(rule, description, variable_keywords)
            results.append({
                "rule": rule,
                "generated_description": description,
                **result
            })
            recalls.append(result["recall"])  # Extract recall for the separate list
            f1_scores.append(result["f1_score"])  # Extract F1 score for the separate list
    
    return results, recalls, f1_scores
```

```python
def plot_manual_means_with_seaborn(recall_scores, f1_scores, labels, title='gpt2'):
    """
    Plot the mean ROUGE Recall and F1 scores with 95% confidence intervals using Seaborn.

    Parameters:
    - recall_scores (list of lists): Nested list of ROUGE recall scores for each label.
    - f1_scores (list of lists): Nested list of ROUGE F1 scores for each label.
    - labels (list): A list of labels corresponding to each score group.
    - bars_points (str): 'bars' for bar plot or 'points' for point plot.
    """
    # Prepare data in a long format suitable for seaborn
    data = {
        "Label": [],
        "Score Type": [],
        "Score": []
    }

    for label, recall, f1 in zip(labels, recall_scores, f1_scores):
        # Skip this label if either recall or f1 has None values
        if recall is None or f1 is None:
            print(f"Skipping label {label} due to None values in recall or F1 scores")
            continue
        
        # Append Recall scores
        data["Label"].extend([label] * len(recall))
        data["Score Type"].extend(["Recall"] * len(recall))
        data["Score"].extend(recall)
        
        # Append F1 scores
        data["Label"].extend([label] * len(f1))
        data["Score Type"].extend(["F1"] * len(f1))
        data["Score"].extend(f1)

    # Convert to DataFrame for Seaborn
    df = pd.DataFrame(data)

    #plt.figure(figsize=(12, 7))
    plt.figure(figsize=(13, 5))

    
    sns.barplot(data=df, x="Label", y="Score", hue="Score Type", errorbar=('ci', 95), palette = ["#4682B4", "#CD5C5C"])

    # Labeling and formatting
    plt.ylabel('Mean Manual Metric Scores', fontsize=20)
    plt.xlabel("") 
    plt.title(f"{title}", fontsize=24)
    plt.ylim(0, 1)  # Assuming scores are between 0 and 1
    plt.grid(axis='y')
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=20)
    plt.legend(title="Score Type", fontsize=20, title_fontsize=20)
    plt.show()
```
